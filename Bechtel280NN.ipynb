```python
#First, let's split that CSV for testing and trial
import pandas as pd
import numpy as np
import keras as k
from keras.callbacks import ModelCheckpoint
from keras.models import Sequential
from keras.layers import Dense, Activation, Flatten
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error 
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt

import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings 
warnings.filterwarnings('ignore')
warnings.filterwarnings('ignore', category=DeprecationWarning)


```

```python
#first set
df = pd.read_csv('out.csv')
df['split'] = np.random.randn(df.shape[0], 1)

msk = np.random.rand(len(df)) <= 0.7

train = df[msk]
test = df[~msk]

x_train = train.iloc[:,1]
y_train = train.iloc[:,2]

y_test = test.iloc[:,2]

#second set
df2 = pd.read_csv('out2.csv')
df2['split'] = np.random.randn(df2.shape[0], 1)

msk2 = np.random.rand(len(df2)) <= 0.7

train2 = df2[msk2]
test2 = df2[~msk2]

x2_train = train2.iloc[:,1]
y2_train = train2.iloc[:,2]

y2_test = test2.iloc[:,2]

# third set
df3 = pd.read_csv('out3.csv')
df3['split'] = np.random.randn(df3.shape[0], 1)

msk3 = np.random.rand(len(df3)) <= 0.7

train3 = df3[msk3]
test3 = df3[~msk3]

x3_train = train3.iloc[:,1]
y3_train = train3.iloc[:,2]

y3_test = test3.iloc[:,2]

#fourthset
df4 = pd.read_csv('out4.csv')
df4['split'] = np.random.randn(df4.shape[0], 1)

msk4 = np.random.rand(len(df4)) <= 0.7

train4 = df4[msk4]
test4 = df4[~msk4]

x4_train = train4.iloc[:,1]
y4_train = train4.iloc[:,2]

y4_test = test4.iloc[:,2]
```

```python

model = k.Sequential()

model.add(k.layers.Dense(units = 1, activation = 'linear', input_shape=[1]))
model.add(k.layers.Dense(units = 32, activation = 'relu'))
model.add(k.layers.Dense(units = 32, activation = 'relu'))
model.add(k.layers.Dense(units = 32, activation = 'relu'))
model.add(k.layers.Dense(units = 32, activation = 'relu'))
model.add(k.layers.Dense(units = 1, activation = 'linear'))
model.compile(loss='mse', optimizer="adam")

model.summary()
```

```python
#First
model.fit( x_train, y_train, epochs=1000, verbose=100)

y_predicted = model.predict(x_train)

#Second
model.fit( x2_train, y2_train, epochs=1000, verbose=100)

y2_predicted = model.predict(x2_train)

#Third
model.fit( x3_train, y3_train, epochs=1000, verbose=100)

y3_predicted = model.predict(x3_train)

#Fourth
model.fit( x4_train, y4_train, epochs=1000, verbose=100)

y4_predicted = model.predict(x4_train)

```
